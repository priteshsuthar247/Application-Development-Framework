```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
originalDataset = pd.read_csv("dataset.csv")
originalDataset
```

```python
originalDataset.describe()
```

```python
originalDataset.shape
```

```python
nullDataset = originalDataset.isnull().sum()
nullDataset
```

```python
threshold = 0.2
droppedColumns = nullDataset[nullDataset > threshold * len(nullDataset)].index
droppedColumns
```

```python
cleansedDataset = originalDataset.drop(droppedColumns,axis=1)
cleansedDataset
```

```python
cleansedDataset.head(10)
```

```python
cleansedDataset['num_voted_users'].fillna(
    cleansedDataset['num_voted_users'].mean())
```

```python
cleansedDataset.fillna(cleansedDataset.mean(numeric_only=True),inplace = True)
cleansedDataset.fillna(cleansedDataset.select_dtypes('object').mode().iloc[0],inplace=True)
cleansedDataset.select_dtypes('object').mode()
```

```python
cleansedDataset.isnull().sum()
```

```python
cleansedDataset.info()
```

```python
cleansedDataset.describe()
```

```python
top_director = originalDataset['director_name'].value_counts().idxmax()
originalDataset['director_name'].value_counts()
```

```python
top_director
```

```python
cleansedDataset.groupby('genres').agg({'imdb_score':max})
```

```python
cleansedDataset[['movie_title','imdb_score']].sort_values(by='imdb_score',ascending=False).head(5)
```

```python
plt.hist(cleansedDataset['imdb_score'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
sns.histplot(cleansedDataset['imdb_score'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
genre = cleansedDataset['genres'].value_counts().head(10)
sns.barplot(data=genre,legend='full')
plt.xticks(rotation=90)
plt.show()
```
